## Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Features

- `product_id` - Product ID

- `product_name` - Name of the Product
- `category` - Category of the Product
- `discounted_price` - Discounted Price of the Product
- `actual_price` - Actual Price of the Product
- `discount_percentage` - Percentage of Discount for the Product
- `rating` - Rating of the Product
- `rating_count` - Number of people who voted for the Amazon rating
- `about_product` - Description about the Product
- `user_id` - ID of the user who wrote review for the Product
- `user_name` - Name of the user who wrote review for the Product
- `review_id` - ID of the user review
- `review_title` - Short review
- `review_content` - Long review
- `img_link` - Image Link of the Product
- `product_link` - Official Website Link of the Product

```python
data = pd.read_csv('data/amazon.csv')
data.head()
```

We will delete some of the columns that are not required for our analysis:

* product_name	

* user_name

* discount_percentage
* about_product	

```python
data = data.drop(columns = ['product_name', 'discount_percentage', 'about_product', 'user_name'])
data.dropna(inplace=True)
data.head()
```

We only need to keep the largest category for each product.

```python
def extractCategory(text):
    category = text.split("|")[0]
    return category

data['category'] = data['category'].apply(extractCategory)
data.head()
```

Remove the first character ₹ for prices.

```python
def removeSign(text):
    return text.replace("₹", "")

data['actual_price'] = data['actual_price'].apply(removeSign)
data['discounted_price'] = data['discounted_price'].apply(removeSign)
data.head()
```

Convert the numerical data to integer format.

```python
data['discounted_price'] = data['discounted_price'].replace({',': ''}, regex=True).astype(float)
data['actual_price'] = data['actual_price'].replace({',': ''}, regex=True).astype(float)
data['rating'] = data['rating'][data['rating'] != '|'].astype(float)
data['rating_count'] = data['rating_count'].replace({',': ''}, regex=True).astype(int)

data.dropna(inplace=True)
data.head()
```

For those products which have review count less than 10, we will remove them from our dataset.

```python
data = data[data.rating_count >= 10]

data.shape
```

```python
data.to_csv("data/data.csv")
```
